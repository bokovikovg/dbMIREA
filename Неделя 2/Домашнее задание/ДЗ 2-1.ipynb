# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql 
select tracks.Name 
from tracks join invoice_items 
                on tracks.TrackId = invoice_items.TrackId 
            join invoices 
                on invoice_items.InvoiceId = invoices.InvoiceId 
            join customers 
                on customers.CustomerId = invoices.InvoiceId 
where customers.CustomerId = 
        (select customers.CustomerId 
            from customers 
                 join employees 
                on customers.SupportRepId = employees.EmployeeId 
            where employees.HireDate > '2002-09-01')
```

## Задание 2
Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql 
select playlists.Name , round(sum(tracks.UnitPrice),-1) as sum_1 , round(sum(tracks.Milliseconds),-1) as sum_2 
from playlists 
    left join playlist_track 
        on playlists.PlaylistId = playlist_track.PlaylistId 
    left join tracks 
        on playlist_track.TrackId = tracks.TrackId 
group by playlists.Name
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql
select a.name, g.name as 'genre', count(t.TrackId) as 'number of songs'
from artists a
    join albums al
        on a.ArtistId=al.ArtistId
    join tracks t
        on al.AlbumId=t.AlbumId
    join genres g
        on t.GenreId=g.GenreId
group by a.name
```

## Задание 4
Для каждого артиста найти жанр, по которому он заработал наибольшее количество денег, а также вывести соотношение общей суммы, заработанной артистом по всем другим жанрам к общей сумме заработанных денег.

```python
%%sql
select a.name, g.name
from artists a
    join albums al
        on a.ArtistId=al.ArtistId
    join tracks t
        on al.AlbumId=t.AlbumId
    join genres g
        on t.GenreId=g.GenreId
group by a.name

```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql
select c.Company, count(t.TrackId) as 'number of tracks'
from customers c
    join invoices i
        on c.CustomerId=i.CustomerId
    join invoice_items ii
        on i.InvoiceId=ii.InvoiceId
    join tracks t
        on ii.TrackId=t.TrackId
    join genres g
        on g.GenreId=t.GenreId
group by c.Company
having g.name='Rock'
```
